```python
def ashtonString(s, k):
    #
    # Write your code here.
    #
    
    # create unique substr dictionary first
    substr_order = dict()
    for i in range(1,len(s)+1):
        # i denote substring length
        for j in range(len(s)-i + 1):
            substr = s[j:(j+i)]
            #print(substr)
            substr_order[substr] = 0
    
    # we utilize the fact that substring's substring must also be a substring of original string
    l = 1
    while True:        
        if l==1:
            substr_order_l = dict()
            # note that empty string has the lowest rank 
            substr_order_l[''] = -1
            
            uniq_substr_l = list(set([x[:l] for x in substr_order.keys()]))
            #print(uniq_substr_l)
            # argsort using the first l substring
            order_substr = sorted(range(len(uniq_substr_l)),key=lambda x: uniq_substr_l[x])
            for i, substr in enumerate(order_substr):
                substr_order_l[uniq_substr_l[substr][:l]] = i
            #print(substr_order_l)
        else:
            # calculate current l's substr_order using substr_order_l from last l
            substr_order_new_l = dict()
            substr_order_new_l[''] = -1
            uniq_substr_new_l = list(set([x[:l] for x in substr_order.keys()]))
            # get the sorting key for all [:new l ]substring  and sort them                     
            old_l = int(l/2)
            #print(old_l)
            sorted_substr_new_l = sorted(uniq_substr_new_l,key=lambda x: (substr_order_l[x[:old_l]], substr_order_l[x[old_l:l]]  ))
            
            for i, substr in enumerate(sorted_substr_new_l):
                substr_order_new_l[substr] = i
            substr_order_l = substr_order_new_l
        if (l>=len(s)):
            break
        else:
            l = l * 2
    
    substr_order_reverse_dict = {substr_final_order:substr for substr, substr_final_order in substr_order_l.items() if substr_final_order >=0}
    #print(substr_order_reverse_dict)
    
    idx = 0
    i = -1
    while idx + len(substr_order_reverse_dict[i+1]) < k:
        idx = idx + len(substr_order_reverse_dict[i+1])
        i += 1
    output = substr_order_reverse_dict[i+1][k - idx - 1]
    return(output)
        
```
