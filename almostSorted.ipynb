```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the almostSorted function below.
def almostSorted(arr):
    adj_inversion = [] # we define adjacent inversion as i's that arr[i] > arr[i+1]
    for i in range(0, len(arr)-1):
        if arr[i] > arr[i+1]:
            adj_inversion.append(i)
    
    # if adj_inversion is empty then it is sorted 
    if len(adj_inversion) == 0 :
        print("yes")
        return
    
    # since swap takes priority over reversion, we evaluate swap first 
    if len(adj_inversion) == 2:
        first_i = adj_inversion[0]
        second_i = adj_inversion[1] + 1
        
        # evaluate the possibility of swapping first_i and second_i 
        
        swap_possible = 1
        
        if first_i > 0 and ( arr[second_i ] < arr[first_i - 1] ):
            swap_possible = 0
        if second_i < len(arr) - 1 and arr[second_i + 1] < arr[first_i]:
            swap_possible = 0
        
        if swap_possible == 1:
            print("yes")
            print("swap " + str(first_i + 1) + " " + str( second_i + 1))
            return
    elif len(adj_inversion) == 1:
        first_i = adj_inversion[0]
        second_i = adj_inversion[0] + 1
        swap_possible = 1
        
        if first_i > 0 and ( arr[second_i ] < arr[first_i - 1] ):
            swap_possible = 0
        if second_i < len(arr) - 1 and arr[second_i + 1] < arr[first_i]:
            swap_possible = 0
        
        if swap_possible == 1:
            print("yes")
            print("swap " + str(first_i + 1) + " " + str( second_i + 1))
            return
    
    # Lastly we evaluate reverse, note that we must have at least 2 adj inversions 
    if len(adj_inversion) >= 2:
        # we have all the adj invesion already, first make sure that between first and last index, everything in the middle is in reverse order arr[i]>= arr[i+1]        
        last_i = max(adj_inversion) + 1
        first_i = min(adj_inversion)
        
        reverse_possible = 1
        for i in range(first_i, last_i):
            if arr[i] < arr[i+1]:
                reverse_possible = 0 
        
        if reverse_possible == 1:
            print("yes")
            print("reverse " + str(first_i + 1) + " " + str( last_i + 1))
            return
                

    
    # we've traverse all cases, if we get here, its a No
    print("no")
```
